### 8.0 Table Lookup

```python
L_student = [
    {'student_no': '00001', 'name': 'Thomson Yeung', 'gender': 'M'} ,
    {'student_no': '00002', 'name': 'Alice Wong', 'gender': 'F'} ,
    {'student_no': '00003', 'name': 'Monica Lee', 'gender': 'F'} ,
    {'student_no': '00004', 'name': 'Tom Lee', 'gender': 'U'} ,
]
L_gender = [
    {'gender': 'M', 'gender_name': 'Male'} ,
    {'gender': 'F', 'gender_name': 'Femle'}
]
L_gender2 = [
    {'gender_id': 'M', 'gender_name': 'Male'} ,
    {'gender_id': 'F', 'gender_name': 'Femle'}
]
```

```python
df_student=pd.DataFrame.from_records(L_student)
df_gender=pd.DataFrame.from_records(L_gender)
df_gender2=pd.DataFrame.from_records(L_gender2)
df_student_rollback = df_student.copy()
```

```python
pd.merge(df_student, df_gender, on='gender')
# select * from df_student s, df_gender g where s.gender=g.gender
```

```python
pd.merge(df_student, df_gender2, left_on='gender', right_on='gender_id')
# select * from df_student s, df_gender2 g where s.gender=g.gender_id
```

```python
pd.merge(df_student, df_gender, how='left', on='gender')
# select * from df_student s, df_gender g where s.gender *= g.gender
```

```python
pd.merge(df_student, df_gender2, how='left', left_on='gender', right_on='gender_id')
# select * from df_student s, df_gender2 g where s.gender *= g.gender_id
```
